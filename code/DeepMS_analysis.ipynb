```python
# Optional: combine those signatures with cosine similarity > 0.8
from numpy import dot
from numpy.linalg import norm

def cosine(a,b):
    return dot(a, b)/(norm(a)*norm(b))
    
def combine_similar_columns(df):
    comb_ls = []
    latent_n = df.shape[1]
    for i in range(1, latent_n+1):
        for j in range(i+1, latent_n+1):
            # print(i,j)
            cos_sim = cosine(df[str(i)].tolist(),df[str(j)].tolist())
            if cos_sim >= 0.8:
                comb_ls.append([i,j])
    # remove those columns
    col_to_remove = []
    for pair in comb_ls:
        col_to_remove.append(pair[1])
    col_to_remove = [str(i) for i in set(col_to_remove)]
    df_out = df.drop(col_to_remove, axis=1)
    return(comb_ls, col_to_remove, df_out)

comb_ls, col_to_remove, brca_lat_8_distinct = combine_similar_columns(brca_lat_8)
print(comb_ls, col_to_remove)
```

# Cosine similarity

```python
import pandas as pd
dir = '/Genomics/pritykinlab/zhiyue/513/DeepMS/'
lat = pd.read_csv(dir + 'DeepMS_TCGA-ALL_counts_preproc.csv_8_32_0.0001_0.0/Latents_TCGA-ALL_counts_preproc.csv_8_32_0.0001_0.0.tsv', sep = '\t', index_col = 0)
lat
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics.pairwise import cosine_similarity
def cos_sim_heatmap(n):
    lat = pd.read_csv(dir + 'DeepMS_TCGA-ALL_counts_preproc.csv_' + str(n) + '_32_0.0001_0.0/Latents_TCGA-ALL_counts_preproc.csv_' + str(n) + '_32_0.0001_0.0.tsv', sep = '\t', index_col = 0)
    n = lat.shape[1]
    arr = cosine_similarity(lat.T)
    corr = pd.DataFrame(arr, columns = np.arange(1,n+1), index = np.arange(1,n+1) )
    
    # Getting the Upper Triangle of the co-relation matrix
    matrix = np.triu(corr)
    matrix[np.triu_indices(matrix.shape[0], 0)] = np.nan
    
    # using the upper triangle matrix as mask 
    sns.set_theme(rc={'figure.figsize':(10,8.27)}, style='white')
    sns.heatmap(corr, annot=True, mask=matrix, cmap = "vlag")
```

```python
# combine those signatures with cosine similarity > 0.8
from numpy import dot
from numpy.linalg import norm

def cosine(a,b):
    return dot(a, b)/(norm(a)*norm(b))
    
def combine_similar_columns(n):
    df = pd.read_csv(dir + 'DeepMS_TCGA-ALL_counts_preproc.csv_' + str(n) + '_32_0.0001_0.0/Latents_TCGA-ALL_counts_preproc.csv_' + str(n) + '_32_0.0001_0.0.tsv', sep = '\t', index_col = 0)
    comb_ls = []
    latent_n = df.shape[1]
    for i in range(1, latent_n+1):
        for j in range(i+1, latent_n+1):
            # print(i,j)
            cos_sim = cosine(df[str(i)].tolist(),df[str(j)].tolist())
            if cos_sim >= 0.8:
                comb_ls.append([i,j])
    # remove those columns
    col_to_remove = []
    for pair in comb_ls:
        col_to_remove.append(pair[1])
    col_to_remove = [str(i) for i in set(col_to_remove)]
    df_out = df.drop(col_to_remove, axis=1)
    print(comb_ls, col_to_remove)
    return(df_out)

distinct = combine_similar_columns(8)
distinct
```
